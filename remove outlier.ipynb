```python
import pandas as pd
import glob
import os
import numpy as np
import pandas as pd
from unidecode import unidecode
import re
```

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    # print(df_in[(df_in[col_name]>fence_high) | (df_in[col_name]<fence_low)].index)
    # print(q1,q3)
    # print(fence_low)
    # print(fence_high)
    # df_indf_in[(df_in[col_name]>fence_high) | (df_in[col_name]<fence_low)].index,inplace=True
    df_in.drop(df_in[df_in[col_name]>fence_high].index,inplace=True)
    df_in.drop(df_in[df_in[col_name]<fence_low].index,inplace=True)
```

```python
directory = "csv"
for filename in os.listdir(directory):
    f_i = f"csv/{filename}"
    df =pd.read_csv(f_i)
    df=df[df.total_cost != 'توافقی']
    df.loc[:,'total_cost']= df['total_cost'].astype(str).str.replace("تومان", "")
    df.loc[:,'price_per_meter'] = df['price_per_meter'].astype(str).str.replace("تومان", "")
    df.loc[:,'price_per_meter'] = pd.to_numeric(df['price_per_meter'].astype(str).str.replace("٬", "").apply(unidecode), errors='coerce')

    
    remove_outlier(df,'price_per_meter')
    f_o=f"outlier_remove/{filename}"
    df.to_csv(f_o)
```

```python
files = os.path.join("outlier_remove", "*.csv")
files = glob.glob(files)
df1 = pd.concat(map(pd.read_csv, files), ignore_index=True)
df1.to_csv('clean/divar_out_remove.csv')
df = pd.read_csv('clean/divar_out_remove.csv', encoding="utf-8")
```
